# Realizando consultas

```python
from connect import db
from bson.objectid import ObjectId

resultados = db.libros

# Consultar todos los documentos
for doc in resultados.find():
    print(doc)
```

```python
# consulta por autor
for doc in resultados.find({"autor":"Paulo Coelho"}):
    print(doc['titulo'], "-", doc['año_publicacion'])
```

```python
# Cosulta muestre libros mayores a año 1960
for doc in resultados.find({"año_publicacion": {"$gt": 1960}}):
    print(doc['titulo'], " - ",doc['autor'] , "-", doc['año_publicacion'])
```

```python
# Listar libros de ciencia ficcion
for doc in resultados.find({"Categoria":"Ciencia ficción"}):
    print(doc['titulo'], "-", doc['año_publicacion'])
```

```python
# libros con stock mayor a 2
for doc in resultados.find({"stock": {"$gte": 2}}):
    print(doc['titulo'], "-", doc['año_publicacion'])
```
