# Update documents

```python
from connect import db
from pymongo import UpdateOne

modificar = db.libros

# Estoy agregando 10 unidades al stock del libro "1984"
res = modificar.update_one(
    {"titulo": "1984"},
    {"$set": {"stock": 10}} # "set" para actulizar y "inc" para incrementar
)
print(f"Documentos modificados: {res.modified_count}")
print('CAmbios', res.matched_count, 'Total', res.modified_count)
```

```python
# catulizar titulo de un libro
res = modificar.update_one( 
    {"titulo": "1984"},
    {"$set": {"titulo": "La bella y la bestia"}}
)
```

```python
for doc in modificar.find({"titulo": "La bella y la bestia"}):
    print('Cambios', res.matched_count, 'Total', res.modified_count)
```

```python
# Actualizacio  usando Updatate Many
res2 = modificar.update_many(
    {},
    {"$set": {"disponible": True}}
)
print ('Cambios', res2.matched_count, 'Total', res2.modified_count)
```

```python
for doc in modificar.find({"disponible": True}):
    print(doc['titulo'], "- Disponible: ", doc['disponible'])
```

```python
# bulk write
operations = [
    UpdateOne(
        {"titulo": "La bella y la bestia"},
        {"$set": {"stock": 6}}
    )
]
modificar.bulk_write(operations)
```
