# Update documents

```python
from connect import db
from pymongo import UpdateOne

modificar = db.libros

# Estoy agregando 10 unidades al stock del libro "1984"
res = modificar.update_one(
    {"titulo": "1984"},
    {"$set": {"stock": 10}} # "set" para actulizar y "inc" para incrementar
)
print(f"Documentos modificados: {res.modified_count}")
print('CAmbios', res.matched_count, 'Total', res.modified_count)
```
